In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

import pandas as pd
df= pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
df.head(10)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
list(df.columns)
```

```python
pivot_df = df.pivot_table(index='sales_channel', values=['total_claim_amount'], aggfunc='sum')
```

```python
pivot_df
```

```python
pivot_df.round(2)
```

```python
'''To analyze the data, I considered the total amount 
of claims per sales channel, since they represent output from the insurance company.
We can conclude that Agent sales channel is the less significant in 
company's revenues, since it represents most of the amount claimed by customers.
On the other hand, web sales channel was more effective in relation with the company'''s revenues.
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_df_1 = df.pivot_table(index='gender', columns='education', values=['customer_lifetime_value'], aggfunc='sum')
```

```python
pivot_df_1
```

```python
pivot_df_1.round(2)
```

```python
'''We can conclude that females with Bachelor degree have more value than males. 
Also, in the same order of ideas, females in college and Doctor level has slightly less value than males.
Finally, females with High School or below educational level, just as females with master degree,
represent more value than males as customers through time prospection.'''
```
